```python
#Markov analysis - Full file opening although code should be cleaned
import re
def create_list(word_file):
    '''Good version of a function that downloads a text file into a list'''
    fin = open(word_file, encoding='utf-8', errors='ignore')
    lines = fin.readlines()
    word_list =[]
    for line in lines:
        word = line.split()
        #word = re.split('\W+', line)  #to remove punctation
        word_list.extend(word)
    #for i in range(len(word_list)):
    #    word_list[i] = word_list[i].casefold()
    for element in word_list:
        if element == '':
            word_list.remove(element)
    return word_list
```

```python
def create_Markov_dictionnary(file):
    dico = dict()
    word_list = create_list(file)
    for i in range(len(word_list)):
        if (word_list[i], word_list[i+1]) in dico:
            if word_list[i+2] not in dico[(word_list[i], word_list[i+1])]:
                dico[(word_list[i], word_list[i+1])].append(word_list[i+2])
        else:
            dico[(word_list[i], word_list[i+1])] = [word_list[i+2]]
    return dico
```

```python
def create_Markov_dictionnary_n(file,n):
    dico = dict()
    word_list = create_list(file)
    for i in range(len(word_list)-n):
        tuple_n = tuple()
        for j in range(n):
            tuple_n = tuple_n + (word_list[i+j],)
        if tuple_n in dico:
            if word_list[i+n] not in dico[tuple_n]:
                dico[tuple_n].append(word_list[i+n])
        else:
            dico[tuple_n] = [word_list[i+n]]
    return dico
```

```python
create_Markov_dictionnary_n('oldmanandthesea.txt', 3)
```

```python
import random
def generate_Markov_poetry(file, n, num_words=100):
    '''creates word_list and dico with possible Markov continuations'''
    dico = dict()
    word_list = create_list(file)
    for i in range(len(word_list)-n):
        tuple_n = tuple()
        for j in range(n):
            tuple_n = tuple_n + (word_list[i+j],)
        if tuple_n in dico:
            if word_list[i+n] not in dico[tuple_n]:
                dico[tuple_n].append(word_list[i+n])
        else:
            dico[tuple_n] = [word_list[i+n]]
            
    '''start random generation'''
    poetry = str()
    
    prefix = random.choice(list(dico.keys()))
    for j in range(n):
        poetry = poetry + ' ' + prefix[j]
    for i in range(num_words):
        if prefix in dico:
            prefix = prefix[1:] + (random.choice(dico[prefix]),)
        else:
            print("Error")
        poetry = poetry + " " + prefix[1]
    print(poetry)
        
    
```
